```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels
import statsmodels.api as sm
import sklearn
```

CONVERT DATA TYPE FROM OBJECT TO INT

EDA ( EXPLORATORY DATA ANALYSIS)

```python
print ("Total Sales:",df['Total Sales'].sum())
```

TOP PERFORMING PRODUCT

```python
category_sales = df.groupby('Product')['Total Sales'].sum().sort_values(ascending=False)
print(category_sales)
```

TOP PERFORMING CUSTOMER NAME

```python
top_customer = df.groupby('Customer Name')['Total Sales'].sum().sort_values(ascending=False)
print(top_customer)
```

CUSTOMER LOCATION WISE ANALYSIS

```python
region_sales = df.groupby('Customer Location')['Total Sales'].sum().sort_values(ascending=False)
print(region_sales)
```

PRODUCT Vs TOTAL SALES LINE CHARTS

```python
plt.plot(df['Product'], df['Total Sales'], marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

CATEGORY Vs TOTAL SALES BAR CHARTS

```python
plt.bar(df['Category'], df['Total Sales'])
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

CUSTOMER LOCATION Vs TOTAL SALES analysis

```python

plt.barh(df['Customer Location'], df['Total Sales'])
plt.title("Sales by Customer Location")
plt.show()
```

category wise total sales share analysis

```python
plt.pie(df['Total Sales'], labels=df['Category'], autopct='%1.1f%%', startangle=90)
plt.title("Category-wise Total Sales Share")
plt.show()
```

```python
plt.scatter(df['Product'], df['Quantity'])
plt.title("Product vs Quantity")
plt.xlabel("Productt")
plt.ylabel("Quantity")
plt.show()
```
